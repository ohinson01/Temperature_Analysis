```python
# Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns; sns.set()
```

```python
# Read both CSV Files, also creates DataFrames
jan = pd.read_csv('jan_temp_values.csv')
feb = pd.read_csv('feb_temp_values.csv')

# Print January DataFrame
jan
```

```python
# Print February DataFrame
feb
```

```python
# Create new DataFrame by appending February data to January data, ignoring indices
jan_feb = pd.concat([jan, feb], ignore_index = True)

# Print new DataFrame
jan_feb
```

```python
# Calculate mean and sum of Low column before filling missing data
mean_before = np.nanmean(jan_feb['Low'])
sum_before = np.nansum(jan_feb['Low'])

# Calculate mean and sum of Low column after filling missing data
fill = jan_feb['Low'].fillna(32)
mean_after = np.mean(fill)
sum_after = np.sum(fill)

# Print results
print('Low Column Results:')
print('-------------------')
print('Mean before filling missing data: ', mean_before)
print('Sum before filling missing data: ', sum_before)
print()
print('Mean after filling missing data: ', mean_after)
print('Sum after filling missing data: ', sum_after)
```

```python
# Calculate mean and sum of High column before filling missing data
mean_before = np.nanmean(jan_feb['High'])
sum_before = np.nansum(jan_feb['High'])

# Calculate mean and sum of High column after filling missing data
fill = jan_feb['High'].fillna(49)
mean_after = np.mean(fill)
sum_after = np.sum(fill)

# Print results
print('High Column Results:')
print('--------------------')
print('Mean before filling missing data: ', mean_before)
print('Sum before filling missing data: ', sum_before)
print()
print('Mean after filling missing data: ', mean_after)
print('Sum after filling missing data: ', sum_after)
```

```python
# Check Low column for missing data
low_na = jan_feb['Low'].isnull()

# Print results
print(low_na)
```

```python
# Check High column for existing data
high_exist = jan_feb['High'].notnull()

# Print results
print(high_exist)
```

```python
# Fill missing Low value 
fill = jan_feb['Low'].ffill()
# Print result
print(fill)
```

```python
# Fill High column
fill = jan_feb['High'].bfill()

# Print result
print(fill)
```

```python
# Create scatter plot
# X-axis as 'Date', Y-axis as 'Low Temp'
jan_feb.plot.scatter(x = 'Date', y = 'Low')
```

```python
# Create linear regression
# X-axis as 'Date', Y-axis as 'High Temp'
jan_feb[]
sns.regplot(x = 'Date', y = 'High', data = jan_feb)
```

```python
# Create linear history graph

# Fill missing Low value 
low_fill = jan_feb['Low'].ffill()

# Fill High column
high_fill = jan_feb['High'].bfill()

# Store columns in variables
x = jan_feb['Date']
y1 = low_fill
y2 = high_fill
y3 = (y1 + y2) / 2 # Calculate mean of Temperatures

# PLot Linear History Graph
plt.figure(figsize = (15, 4))
plt.title('Linear History Graph for Low Temps, High Temps, and Avg Temps', fontdict = {'fontsize' : 14})
plt.tick_params(axis = 'x', labelrotation = 90)
plt.plot(x, y1, linewidth = 4, label = 'Low Temp')
plt.plot(x, y2, linewidth = 4, label = 'High Temp')
plt.plot(x, y3, linewidth = 4, label = 'Avg Temps')
plt.xlabel('Date')
plt.ylabel('Temperatures (F)')
plt.legend()
```
